```python
def limpiar_fecha(x): 
    fecha_string1 = x.split("_")[1]
    fecha_date = datetime.datetime.strptime(fecha_string1, '%Y%m%d')
    return fecha_date

def limpiar_handle(x): 
    handle_sr = x.split("_")[0]
    return handle_sr
```

```python
df_tw["handle"] = df_tw["unique_id"].apply(limpiar_handle)
df_tw["fecha"] = df_tw["unique_id"].apply(limpiar_fecha)
```
